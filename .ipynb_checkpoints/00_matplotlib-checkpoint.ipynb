```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1, 5, 100)
y0 = x
y1 = x * x
y2 = x * x * x
y3 = x * np.log(x)
y4 = np.sqrt(x)

plt.plot(x, y0)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)

plt.legend([
    'y = x',
    'y = $x^2$',
    'y = $x^3$',
    'y = x * $log(x)$',
    'y = $\sqrt{x}$'
])

plt.xlabel('x')
plt.xlabel('y')

plt.show()
```
